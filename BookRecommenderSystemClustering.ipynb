```python
x = ratings['user-id'].value_counts()>200
```

```python
x[x].shape
```

```python
y= x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user-id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books =  ratings.merge(books , on = 'ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns ={'rating':'number of ratings'} , inplace =True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating , on = 'title')
```

```python
final_rating
```

```python
final_rating.shape
```

```python
new_final_rating = final_rating[final_rating['number of ratings']>=50]
```

```python
new_final_rating.shape
```

```python
new_final_rating.drop_duplicates(['user-id' , 'title'] , inplace = True)
```

```python
new_final_rating.shape
```

```python
book_pivot = new_final_rating.pivot_table(columns = 'user-id' , index = 'title' , values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0 , inplace = True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors( algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distances , suggestions = model.kneighbors (book_pivot.iloc[237 , :].values.reshape(1, -1) , n_neighbors = 6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
def recommend_books(book_name):
  book_id = np.where(book_pivot.index == book_name)[0][0]
  distances , suggestions = model.kneighbors (book_pivot.iloc[book_id , :].values.reshape(1, -1) , n_neighbors = 6)

  for i in range(len(suggestions)):
    if i ==0:
      print("The sugessions for" , book_name , "are :")
    if not i :
      print(book_pivot.index[suggestions[i]])
```

```python
recommend_books('The Cradle Will Fall')
```
